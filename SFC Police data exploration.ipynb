# Exploration of the SFC Police Data 

Exploring the open data from Sant Franscisco police departement



```python
import pandas as pd
```

```python
df = pd.read_csv('./data/Police_Department_Incidents.csv',low_memory=False)
```

```python
df.head(n=10)
```

Explore the effectiveness of the police by seen what is the resolution of the crimes.


```python
res = df.groupby('Resolution')['Time'].count()
res.sort_values(ascending=False)
```

Now let's visualize the top 15 crimes of the city. 

```python
from bokeh.charts import Bar
from bokeh.charts.attributes import ColorAttr, CatAttr
from bokeh.io import show, output_notebook

output_notebook()
```

```python
crime = df.groupby('Category')['Time'].agg(['count', 'max'])
crime_top = crime.sort_values('count',ascending=False).head(n=15)
```

```python
p_top_crime = Bar(crime_top,values='count',
                  legend=None,label=CatAttr(columns=['index'], sort=False),
                  color='index',
                  width=900)                  
show(p_top_crime)
```

Now we will list the top crimes by number of arrests.

```python
crime_ar = df[df.arrested].groupby('Category')['Time'].agg(['count', 'max'])
crime_top_ar = crime_ar.sort_values('count',ascending=False).head(n=15)
```

```python
p_top_crime_ar = Bar(crime_top_ar,values='count',
                  legend=None,label=CatAttr(columns=['index'], sort=False),
                  color='index',
                  width=900)                  
show(p_top_crime_ar)
```

We will explore the resolution by crime type.

```python
res_bycrime = df.groupby(['Category', 'Resolution'] )['Time'].count()
res_bycrime.unstack()
```

We can see that there are too many categories for resolution, we understand that when resolution is `NONE` the police was not able to arrest the supect of the crime. In the other cases the police manage well. 

We create a new column to the dataset called `arrested` with binray value that will define if the police manage to arrest the suspect. 

```python
df['arrested'] = df['Resolution'] != 'NONE'
```
